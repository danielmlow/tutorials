```python
# bar plot
x = counts['Crisis'].values
heights = counts['Count'].values


import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Arial"
plt.rcParams["font.size"] = 17
plt.rcParams["figure.figsize"] = (12,8)
plt.bar(x=x,height = heights,
       color='dodgerblue')
plt.xticks(rotation=85, size=15)
plt.ylabel('Amount of conversations')
plt.xlabel('Type of crisis')
plt.tight_layout()
plt.show()
# plt.savefig('./../../../data/ctl/output/tags_proportion.png', dpi=300)
```
