```python
import opensmile
from scipy.io import wavfile
from google.colab import files
import os
import glob
```

```python
# Get access to your google drive. 
from google.colab import drive
drive.mount('/content/drive/')
```

```python
# config: change to your audio types or convert to this
sampling_rate = 16000
bit_rate = '64k'
feature_level = 'func'  # or 'lld' for matrix sequential data
```

```python
input_dir = '/content/drive/My Drive/my_notes/audio_samples/'
output_dir = '/content/drive/My Drive/my_notes/egemaps/'
# audio_files = os.listdir(input_dir)
audio_files = glob.glob(input_dir+'*', recursive=True)
audio_files
```

```python
smile = opensmile.Smile(
		feature_set=opensmile.FeatureSet.eGeMAPSv02, #or path to conf: 'gemaps/eGeMAPSv02.conf'
		feature_level=feature_level,
		verbose=True,
	)

	
vectors = smile.process_files(audio_files)
```

```python
vectors
```

```python
import datetime
ts = datetime.datetime.utcnow().strftime('%y-%m-%dT%H-%M-%S') # so you don't overwrite, and save timestamp
filename = f'egemaps_{ts}.csv'
```

```python

# save to folder
os.makedirs(output_dir, exist_ok=True) #make output dir if not exists
vectors.to_csv(output_dir+filename) #ts is to save timestamp and not override previous versions. 
```
