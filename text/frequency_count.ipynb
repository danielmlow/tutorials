```python
from collections import Counter
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
corpus = [
    'This is the first document.',
    'This is a document, the second document.',
    'And this is the third one.',
    'Is this the first document?',
]
```

```python
%%time
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)
feature_names = vectorizer.get_feature_names_out()
display(pd.DataFrame(X.toarray(), columns = feature_names))
```

```python
#use ngrams
vectorizer2 = CountVectorizer(analyzer='word', ngram_range=(1, 5))
X2 = vectorizer2.fit_transform(corpus)
feature_names = vectorizer2.get_feature_names_out()
df = pd.DataFrame(X2.toarray(), columns = feature_names)
display(df)
```
