```python
import re

def convert_bold_to_html(text):
    """Converts bold markdown to HTML bold tags with proper line breaks and indents."""

    # Split the text into lines to preserve formatting
    lines = text.split('\n')
    converted_lines = []

    for line in lines:
        # Replace **text** with <strong>text</strong>
        converted_line = re.sub(r'\*\*(.*?)\*\*', r'<strong>\1</strong>', line)
        converted_lines.append(converted_line)

    # Join the lines back together, adding HTML line breaks and indentation
    html_text = ""
    for i, line in enumerate(converted_lines):
        if line.strip():  # Only add indentation/breaks if the line is not empty
            indent_level = len(line) - len(line.lstrip())
            html_text += "    " * indent_level + line.strip()  # Add indentation 
            if i < len(converted_lines) -1:
                html_text += "<br><br>\n"
            
    return html_text

final_result_html = convert_bold_to_html(final_result)
print(final_result_html)
```

```python
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart


def send_email(sender_email, receiver_email, subject, body, smtp_server, smtp_port, login, password):
    # Create the email
    message = MIMEMultipart()
    message['From'] = sender_email
    message['To'] = receiver_email
    message['Subject'] = subject

    # Attach the email body
    message.attach(MIMEText(body, 'html'))

    # Set up the SMTP server
    with smtplib.SMTP(smtp_server, smtp_port) as server:
        server.starttls()  # Secure the connection
        server.login(login, password)  # Login
        server.sendmail(sender_email, receiver_email, message.as_string())  # Send email

# Example usage
sender_email = email_keys.sender_email 
receiver_email = email_keys.receiver_email # change 
subject = f"Resumen del episodio de Huberman: {filename.replace('.txt', '')}"
body = "Este es un mail automatizado:<br><br>" + final_result_html
smtp_server = "smtp.gmail.com"
smtp_port = 587
login = email_keys.login
password = email_keys.password  # Use an App Password for Gmail

send_email(sender_email, receiver_email, subject, body, smtp_server, smtp_port, login, password)
```
